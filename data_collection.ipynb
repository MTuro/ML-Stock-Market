# Data Collection
Steps and code for collecting and saving S&P 500 historical data.

```python
#Imports

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Download historical data for the S&P 500
ticker = '^GSPC'
df = yf.download(ticker, start='2015-01-01', end='2024-12-31')
```

```python
#Preview the first 5 rows
display(df.head())
```

```python
#Plot closing prices
plt.figure(figsize=(14, 5))
plt.plot(df['Close'], label='Closing Price')
plt.title(f'{ticker} Historical Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#save locally
df.to_csv('sp500.csv')
```
